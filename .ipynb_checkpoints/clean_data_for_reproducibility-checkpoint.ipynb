Modify data to make it cleaner for the reproducibility example.

```python
import pandas
```

```python
topic_words = pandas.read_csv("data/comparativewomensmovement_topweightedtopicwords.csv", index_col=0) #read in dataset including top weighted words for each topic
topic_words.columns = topic_words.columns.str.replace('V', "X")
topic_words
```

```python
topic_words.to_csv("data/comparativewomensmovement_topweightedtopicwords.csv")
```

```python

df = pandas.read_csv("data/comparativewomensmovement_dataset_withtopicweights.csv", index_col=0) #read in dataset, including text, metadata, and topic weights



col_list = []
for c in range(1,41):
    col = "X"+str(c)
    new_col = "X"+str(c)
    col_list.append(new_col)
    df[new_col] = df.apply(lambda row: (row[col] * row['word_count']), axis=1)
col_list.append('word_count')
col_list.append('org')

df
```

```python
df.to_csv("data/comparativewomensmovement_dataset_wordcount.csv")
```
